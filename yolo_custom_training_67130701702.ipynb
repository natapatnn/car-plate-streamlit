## Fix data.yaml missing validation set

```python

with open(data_yaml_path, "r") as f:
    data_config = yaml.safe_load(f)

if "val" in data_config:
    val_path = os.path.join(dataset.location, data_config["val"])
    if not os.path.exists(val_path):
        print("âš ï¸ Validation set missing â†’ à¹ƒà¸Šà¹‰ train/images à¹à¸—à¸™")
        data_config["val"] = "train/images"

with open(data_yaml_path, "w") as f:
    yaml.dump(data_config, f)

print("âœ” data.yaml updated")
```

## Train Model 1: YOLO11n

```python
model1 = YOLO("yolov8n.pt")

result1 = model1.train(
    data=data_yaml_path,
    epochs=10,
    imgsz=640,
    batch=16,
    lr0=0.01,
    lrf=0.1,
    patience=20
)

print("âœ” Model 1 Finished")
```

## Train Model 2: YOLO11s

```python
model2 = YOLO("yolov8s.pt")

result2 = model2.train(
    data=data_yaml_path,
    epochs=10,
    imgsz=640,
    batch=16,
    lr0=0.01,
    lrf=0.1,
    patience=20
)

print("âœ” Model 2 Finished")
```
